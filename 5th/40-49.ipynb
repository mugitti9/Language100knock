```python
#40
class Morph:
    def __init__(self, surface, base, pos, pos1):
        self.surface = surface
        self.base = base
        self.pos = pos
        self.pos1 = pos1

dependency=[]
dependency_before=[]
with open("ai.ja.txt.parsed", mode='rt', encoding='utf-8') as f:
    for line in f:
        before=line.split()
        if before!=[]:
            if before!=['EOS']:
                surface=before[0]
                word_class=before[1].split(",")
                if len(word_class)>1:
                    dependency_before.append(Morph(surface, word_class[6], word_class[0], word_class[1]))
            elif (before==['EOS'] and len(dependency_before)!=0):
                dependency.append(dependency_before)
                dependency_before=[]
                
for dependency_one in dependency[1]:
    print(vars(dependency_one))
```
